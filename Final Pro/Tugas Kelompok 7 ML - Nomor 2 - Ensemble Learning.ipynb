#DATA PREPROCESSING

##Data Cleaning

```python
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)

clean_dataset(scaled_dt)
clean_dataset(scaled_ds)
clean_dataset(dt)
clean_dataset(ds)
```

```python
x = (df_train.dtypes == 'object')
object_cols = list(x[x].index)
y = (df_test.dtypes == 'object')
object_cols1 = list(y[y].index)

print("Categorical variables in the datatrain:", object_cols)
print("Categorical variables in the datatest:", object_cols1)
```

##Label Encoder Data

Karena dataset berisi variabel karakter, maka harus menyandikannya menggunakan LabelEncoder

```python
#Label Encoding the object dtypes.
LE=LabelEncoder()
for i in object_cols:
    df_train[i]=df_train[[i]].apply(LE.fit_transform)

for j in object_cols1:
    df_test[j]=df_test[[j]].apply(LE.fit_transform)
    
print("All features are now numerical")
```

##Normalisasi(Feature scaling)

```python
#Creating a copy of data
dt = df_train.copy()
ds = df_test.copy()
#Scaling
scaler = StandardScaler()
scaler.fit(dt)
scaled_dt = pd.DataFrame(scaler.transform(dt),columns= dt.columns )
scaler.fit(ds)
scaled_ds = pd.DataFrame(scaler.transform(ds),columns= ds.columns )
print("All features are now scaled")
```

```python
print(f'Shape of datatrain = {scaled_dt.shape}')
scaled_dt.head(10)
```

```python
scaled_ds = scaled_ds.iloc[25:]
print(f'Shape of datatest = {scaled_ds.shape}')
scaled_ds.head(10)
```

```python
df_train.info()
```

```python
X = scaled_dt.iloc[:, :-1]
y = scaled_dt.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.33, random_state = 69)

print(f'Training set--> X_train shape= {X_train.shape}, y_train shape= {y_train.shape}')
print(f'Holdout set--> X_test shape= {X_test.shape}, y_test shape= {y_test.shape}')
```

##Root Mean Square Error (RMSE) 

```python
models_scores = [] # To store model scores

def rmse(model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    return mean_squared_error(y_test, y_pred, squared= False) # squared= False > returns Root Mean Square Error                  
```

#MODEL REGRESSION

##LINEAR REGRESSION

```python
linear_regression = make_pipeline(LinearRegression())
score = rmse(linear_regression)

models_scores.append(['LinearRegression', score])
print(f'LinearRegression Score= {score}')
```

##LASSO REGRESSION

```python
lasso = make_pipeline(RobustScaler(), Lasso(alpha=0.0005, random_state= 69))

score = rmse(lasso)
models_scores.append(['Lasso', score])
print(f'Lasso Score= {score}')
```

##ELASTICNET REGRESSION

```python
elastic_net = make_pipeline(RobustScaler(), ElasticNet(alpha=0.0005, l1_ratio= .9, random_state= 69))

score = rmse(elastic_net)
models_scores.append(['ElasticNet', score])
print(f'ElasticNet Score= {score}')
```

##KERNEL RIDGE REGRESSION

```python
kernel_ridge= KernelRidge(alpha=0.6, kernel='polynomial', degree=2, coef0=2.5)
score = rmse(kernel_ridge)
models_scores.append(['KernelRidge', score])
print(f'KernelRidge Score= {score}')
```

```python
# Ranking the scores of each model
pd.DataFrame(models_scores).sort_values(by=[1], ascending=True)
```

#ENSEMBLE LEARNING

##BAGGING REGRESSOR

```python
def bagging_predictions(estimator):
    """
    I/P
    estimator: The base estimator from which the ensemble is grown.
    O/P
    br_y_pred: Predictions on test data for the base estimator.
    
    """
    regr = BaggingRegressor(base_estimator=estimator,
                            n_estimators=10,
                            max_samples=1.0,
                            bootstrap=True, # Samples are drawn with replacement
                            n_jobs= n_jobs,
                            random_state=random_state).fit(X_train, y_train)

    br_y_pred = regr.predict(X_test)

    rmse_val = mean_squared_error(y_test, br_y_pred, squared= False) # squared= False > returns Root Mean Square Error   

    print(f'RMSE for base estimator {regr.base_estimator_} = {rmse_val}\n')
    return br_y_pred


predictions = np.column_stack((bagging_predictions(linear_regression),
                              bagging_predictions(lasso),
                              bagging_predictions(elastic_net),
                              bagging_predictions(kernel_ridge)))

print(f"Bagged predictions shape: {predictions.shape}")
```

##FOREST OF RANDOMIZED TREES

###Random Forest Regressor

```python
RandForestRegressor= RandomForestRegressor(n_estimators=3000,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   random_state = random_state)

score = rmse(RandForestRegressor)
models_scores.append(['RandForestRegressor', score])
print(f'RandForestRegressor Score= {score}')
```

###Extra Trees Regressor

```python
ExTreesRegressor= ExtraTreesRegressor(n_estimators=3000,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   random_state = random_state)

score = rmse(ExTreesRegressor)
models_scores.append(['ExTreesRegressor', score])
print(f'ExTreesRegressor Score= {score}')
```

##BOOSTING REGRESSOR

###AddBoost Regressor

```python
bagging_regressor= BaggingRegressor(n_estimators=3000,
                                   random_state = random_state)

score = rmse(bagging_regressor)
models_scores.append(['BaggingRegressor', score])
print(f'BaggingRegressor Score= {score}')
```

###Gradient Boosting Regressor

```python
gradient_boosting_regressor= GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber', random_state = random_state)

score = rmse(gradient_boosting_regressor)
models_scores.append(['GradientBoostingRegressor', score])
print(f'GradientBoostingRegressor Score= {score}')
```

###XGB Regressor

```python
xgb_regressor= xgb.XGBRegressor(colsample_bytree=0.4603, gamma=0.0468, 
                             learning_rate=0.05, max_depth=3, 
                             min_child_weight=1.7817, n_estimators=3000,
                             reg_alpha=0.4640, reg_lambda=0.8571,
                             subsample=0.5213,verbosity=0, nthread = -1, random_state = random_state)
score = rmse(xgb_regressor)
models_scores.append(['XGBRegressor', score])
print(f'XGBRegressor Score= {score}')
```

###ADB Regressor

```python
adb_regressor= AdaBoostRegressor(n_estimators=3000,learning_rate=0.05,loss='linear',
                                 random_state=None)

score = rmse(adb_regressor)
models_scores.append(['ADBRegressor', score])
print(f'ADBRegressor Score= {score}')
```

##Voting Regressor

```python
estimators = [ ('elastic_net', elastic_net), ('kernel_ridge', kernel_ridge),('xgb_regressor', xgb_regressor) ]

VoteRegressor = VotingRegressor(estimators=estimators)

score = rmse(VoteRegressor)
models_scores.append(['VoteRegressor', score])
print(f'VoteRegressor Score= {score}')
```

##Stacking Regressor

```python
estimators = [ ('elastic_net', elastic_net), ('kernel_ridge', kernel_ridge),('xgb_regressor', xgb_regressor) ]

stack = StackingRegressor(estimators=estimators, final_estimator= lasso, cv= 5, n_jobs= n_jobs, passthrough = True)

stack.fit(X_train, y_train)

pred = stack.predict(X_test)

rmse_val = mean_squared_error(y_test, pred, squared= False) # squared= False > returns Root Mean Square Error    
models_scores.append(['Stacking', rmse_val])
print(f'rmse= {rmse_val}')
```

```python
# Ranking the scores of each model
pd.DataFrame(models_scores).sort_values(by=[1], ascending=True)
```
