#Data Preprocessing dan Feature Scaling

Menload Dataset menjadi x dan y

```python
# Memilih cell tertentu dari dataset
X = data.iloc[:, [0, 1, 2]].values
y = data.iloc[:, -1].values
```

Split Dataset menjadi Test dan Train

```python
# Import sklearn untuk mensplit
# Split  menjadi test dan train dengan test_size = 0.2
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

Scaling Dataset

```python
# Feature Scalling untuk mentrain dan mentest set dari variabel indepen untuk di kurangi menjadi value yang lebih kecil
from sklearn.preprocessing import StandardScaler

# Membuat method untuk di scaling
scaler = StandardScaler()

# Menscaling dataset x_train dan x_test
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Algoritma KNN sebagai Model Classifier

```python
# Membuat dan mentrain KNN model dengan training dataset
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

Membandingkan True dan Predict Value

```python
y_test
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

#Evaluasi Model

```python
# Import sklearn matrix untuk evaluasi dan akurasi model dari algoritma KNN
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

Accuracy_score

```python
# Hasil prediksi menggunakan accuracy_score
ac = accuracy_score(y_test,y_pred)
print('Testing Accuracy = {}'.format(ac))
```

Confusion_matrix

```python
# Hasil prediksi menggunakan confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrix = ')
print(cm)
```

Classification_report

```python
# Hasil prediksi menggunakan classification_report
cr = classification_report(y_test,y_pred)
print('classification report = ')
print(cr)
```
