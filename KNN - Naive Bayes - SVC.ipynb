#Data Preprocessing dan Feature Scalling

Menload Dataset menjadi x dan y

```python
# Memilih cell tertentu dari dataset
X = data.iloc[:, [0, 1, 2]].values
y = data.iloc[:, -1].values
```

Split Dataset menjadi Test dan Train

```python
# Import sklearn untuk mensplit
# Split  menjadi test dan train dengan test_size = 0.2
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

Scaling Dataset

```python
# Feature Scalling untuk mentrain dan mentest set dari variabel indepen untuk di kurangi menjadi value yang lebih kecil
from sklearn.preprocessing import StandardScaler

# Membuat method untuk di scaling
scaler = StandardScaler()

# Menscaling dataset x_train dan x_test
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Mencari Parameter untuk menggunakan KNN dan GridSearch

```python
# Import Sklearn untuk KNN dan Gridsearch
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# Method untuk KNN dengan n_neighbours = 5
knn = KNeighborsClassifier(n_neighbors=5)

# Menentukan range
k_range = range(1,31)

# Merthod GridSearch dengan fold = 5 dan roc_auc
param_grid = dict(n_neighbors=k_range)
grid = GridSearchCV(knn, param_grid=param_grid, cv=5, scoring='roc_auc')
print(grid)
```

Menampilkan parameter terbaik di KNN

```python
# print grid_ untuk menampilkan parameter
grid.fit(X_train,y_train)
print(grid.best_score_)
print(grid.best_params_)
print(grid.best_estimator_)
```

# Ulangi Menggunakan SVC

```python
# Import Sklearn untuk SVC
from sklearn.svm import SVC
classifier = SVC(kernel='linear',random_state=0)

# Menentukan Parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001]}
 
gridSVC = GridSearchCV(classifier, param_grid, cv=5, scoring='roc_auc')

# Fitting model
gridSVC.fit(X_train, y_train)

```

Menampilkan Hasil parameter SVC

```python
# Menampilkan hasil
print(gridSVC.best_score_)
print(gridSVC.best_params_)
```
